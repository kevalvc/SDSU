# Question 1: Data Visualization (Results and Code)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Straight Line

x = np.arange(-5.0, 5.0, 0.01)
t0 = 30
t1 = 0.5

y = t0 + (t1 * x)

line = plt.plot(x, y, lw=2)

#plt.ylim(-2,2)
plt.show()
```

```python
# Quadratic Equation

x = np.arange(-100.0, 100.0, 0.01)
t1 = 25
t0 = 20

y = np.square(x - t1) + t0

line = plt.plot(x, y, lw=2)

plt.show()
```

```python
# Log Function

x = np.arange(0.1, 1, 0.01)
        
y1 = -np.log(x)
y2 = -np.log(1-x)

line1 = plt.plot(x, y1, lw=2)
line2 = plt.plot(x, y2, lw=2)

plt.show()
```

```python
# Sigmoid Function

x = np.arange(-10.0, 10.0, 0.1)

y = 1 / (1 + np.exp(-x))

line = plt.plot(x, y, lw=2)

plt.show()
```
